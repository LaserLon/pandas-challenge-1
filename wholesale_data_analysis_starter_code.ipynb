## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['subtotal'] = df['unit_price'] * df['qty']
df.head(1)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
df['shipping_rate'] = (df['unit_weight'] * df['qty']).apply(lambda x: 7 if x > 50 else 10)
df['shipping'] = df['shipping_rate'] * df['unit_weight'] * df['qty']
df.head(1)
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df['total_price'] = (df['subtotal'] + df['shipping']) * 1.0925
df.head(1)
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).

df['line_price'] = (df['unit_cost'] * df['qty'] + df['shipping'])
df.head(1)
```

```python
# Create a column for the profit of each line using line cost and line price
df['line_profit'] = df['total_price'] - df['line_price']
df.head(1)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# Get the top 5 most frequent client_ids
top_indexes = df['client_id'].value_counts().head(5).index.to_list()

# Loop through each client_id and sum the total_price
for client_id in top_indexes:
    total = df[df['client_id'] == client_id]['total_price'].sum()
    print(f"Client ID: \033[1m{client_id}\033[0m, Total Price: ${total:,.2f}")
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
summary_data = []
for client_id in top_indexes:
    filtered_df = df[df['client_id'] == client_id]
    
    qty_sum = filtered_df['qty'].sum()
    shipping_sum = filtered_df['shipping'].sum()
    total_price_sum = filtered_df['total_price'].sum()
    line_profit_sum = filtered_df['line_profit'].sum()
    
    summary_data.append({
        'client_id': client_id,
        'qty': qty_sum,
        'shipping': shipping_sum,
        'total_revenue': total_price_sum,
        'total_profit': line_profit_sum
    })

# Create a summary DataFrame
summary_df = pd.DataFrame(summary_data)
summary_df


    # print(f"Client ID: \033[1m{client_id}\033[0m, Total Units: {total_units}, Total Shipping: ${total_shipping:,.2f}, Total Revenue: ${total_revenue:,.2f}, Total Profit: ${total_profit:,.2f}")
```

```python
# Format the data and rename the columns to names suitable for presentation.
summary_df.rename(columns={'client_id': 'Client', 'qty': 'Total Units','shipping': 'Total Shipping', 'total_revenue': 'Total Revenue','total_profit': 'Total Profit'}, inplace=True)

# Define the money columns. 
summary_df['Total Shipping'] = pd.to_numeric(summary_df['Total Shipping'])
summary_df['Total Revenue'] = pd.to_numeric(summary_df['Total Revenue'])
summary_df['Total Profit'] = pd.to_numeric(summary_df['Total Profit'])
summary_df
# Define a function that converts a dollar amount to millions.
def convert_dollar_amount(amount):
    if pd.notnull(amount):
        return amount / 1000000
    return amount


# Apply the currency_format_millions function to only the money columns. 
summary_df['Total Shipping'] = summary_df['Total Shipping'].apply(convert_dollar_amount)
summary_df['Total Revenue'] = summary_df['Total Revenue'].apply(convert_dollar_amount)
summary_df['Total Profit'] = summary_df['Total Profit'].apply(convert_dollar_amount)


# Rename the columns to reflect the change in the money format. 

summary_df.rename(columns={'Total Shipping': 'Total Shipping (millions)', 
                           'Total Revenue': 'Total Revenue (millions)', 
                           'Total Profit': 'Total Profit (millions)'}, inplace=True)
summary_df
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
final_summary_df = summary_df.sort_values(by='Total Profit (millions)', ascending=False)
final_summary_df
```
